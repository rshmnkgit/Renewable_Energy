```python
import pandas as pd
from pprint import pprint
```

```python
data_path="../Resources/"
solar_data_read=pd.read_csv(data_path+"solar-energy-consumption.csv")
solar_data_read.loc[solar_data_read["Entity"]=="Germany"]
```

```python
hydropower_data_read=pd.read_csv(data_path+"hydropower-consumption.csv")
hydropower_data_read
```

```python
# Solar and Hydropower energy merged
merged_solar_hydro_data= pd.merge(solar_data_read,hydropower_data_read,on=["Entity","Year"],how="outer")
merged_solar_hydro_data
```

```python
wind_data_read=pd.read_csv(data_path+"wind-generation.csv")
wind_data_read
```

```python
merged_solar_hydro_wind_data= pd.merge(merged_solar_hydro_data,wind_data_read,on=["Entity","Year"],how="outer")
merged_solar_hydro_wind_data
```

```python
biofuel_data_read=pd.read_csv(data_path+"biofuel-production.csv")
biofuel_data_read
```

```python
mergedAll_data= pd.merge(merged_solar_hydro_wind_data,biofuel_data_read,on=["Entity","Year"],how="outer")
mergedAll_data
```

```python
#Clean Data
mergedAll_data.drop(columns=["Code_x","Code_y","Code_x","Code_y"],inplace=True)
```

```python
mergedAll_data
```

```python
#make the NaN zero
mergedAll_data.fillna(0,inplace=True)
```

```python
clean_df=mergedAll_data.rename(columns={"Entity":"Country",
                               "Electricity from solar (TWh)":"Solar",
                               "Electricity from hydro (TWh)":"Hydro",
                               "Electricity from wind (TWh)":"Wind",
                               "Biofuels Production - PJ - Total":"Biofuels"})
clean_df
```

```python
clean_df.to_csv("../Resources/energyconsumption.csv",index=False)
```

# Share Electricity from Renewables Energies
How much of our electricity comes from renewables?

```python
share_electricity= pd.read_csv(data_path+"share-electricity-renewables.csv")
# share_electricity.loc[share_electricity["Entity"]=="Germany"]
share_electricity.loc[share_electricity["Entity"]=="Germany"]
```

```python
share_solar= pd.read_csv(data_path+"share-electricity-solar.csv")
share_solar.loc[share_solar["Entity"]=="Germany"]
```

```python
merge_share_electr_solar=pd.merge(share_electricity,share_solar, on=["Entity","Year"], how="outer")
merge_share_electr_solar
```

```python
share_wind= pd.read_csv(data_path+"share-electricity-wind.csv")
share_wind
```

```python
merge_share_electr_solar_wind=pd.merge(merge_share_electr_solar,share_wind, on=["Entity","Year"], how="outer")
merge_share_electr_solar_wind
```

```python
share_hydro= pd.read_csv(data_path+"share-electricity-hydro.csv")
share_hydro
```

```python
data_df=pd.merge(merge_share_electr_solar_wind,share_hydro, on=["Entity","Year"], how="outer")
data_df
```

```python
data_df.drop(columns=["Code_x","Code_y","Code_x","Code_y"],inplace=True)
```

```python
#make the NaN zero
data_df.fillna(0,inplace=True)
```

```python
# data_df.loc[data_df["Entity"]=="Germany"]
data_df
```

```python
clean_perc_df=data_df.rename(columns={"Entity":"Country",
                               "Renewables (% electricity)":"Renewables_Perc",
                               "Solar (% electricity)":"Solar_Perc",
                               "Wind (% electricity)":"Wind_Perc",
                               "Hydro (% electricity)":"Hydro_Perc}"})
clean_perc_df
```

```python
clean_perc_df.to_csv("../Resources/renergyshare.csv",index=False)
```

# Death Rate based on Air Pollution in every 100,000 people
Death rates are measured as the number of premature deaths attributed to outdoor air pollution per
100,000 individuals in a given demographic.

```python
air_pollution=pd.read_csv(data_path+"outdoor-pollution-rates-by-age.csv")
air_pollution
```

```python
air_pollution.fillna(0,inplace=True)
```

```python
clear_data_air=air_pollution.rename(columns={"Entity":"Country",
                             "Death rate – Outdoor air pollution (Under-5s) (IHME)":"Death Rate Under_5",
                             "Death rate – Outdoor air pollution (5-14 years) (IHME)":"Death Rate 5_14_Years",
                             "Death rate – Outdoor air pollution (15-49 years) (IHME)":"Death Rate 15_49_Years",
                             "Death rate – Outdoor air pollution (50-69 years) (IHME)":"Death Rate 50_69_Years",
                             "Death rate – Outdoor air pollution (70+ years) (IHME)":"Death Rate Over_70"})
clear_data_air.drop(columns=["Code"],inplace=True)
```

```python
clear_data_air
```

```python
clear_data_air.to_csv("../Resources/airpoleffects.csv",index=False)
```

# Connection to MongoDB

```python
# Define the Database in Mongo
db=client.renewable_energy

#Query all resources
db=client['renewable_energy']
name_airpol=db['airpollution']
```
